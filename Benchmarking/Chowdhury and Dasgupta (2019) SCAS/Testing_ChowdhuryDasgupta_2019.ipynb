# This notebook compares the SCAS calculated from the Chowdhury and Dasgupta (2019) supplement with PySulfSat

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import PySulfSat as ss
pd.options.display.max_columns = None
```

```python
df=pd.read_excel('Excel_Supplement_CD2019.xlsx', 
sheet_name='PySulfSat')
T_K=df['T (ÂºC)']+273.15
```

```python
C_calc=ss.calculate_CD2019_SCAS(df=df, T_K=T_K)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10,4))
ax1.plot(C_calc['SCAS6_ppm'], df['SCAS (ppm)'], 'or')
ax1.set_xlabel('SCAS Python (ppm)')
ax1.set_ylabel('SCAS excel (ppm)')

ax2.plot(C_calc['SCAS6_ppm'], 100-100*C_calc['SCAS6_ppm']/df['SCAS (ppm)'], 'or')
ax2.ticklabel_format(useOffset=False)
ax2.set_ylabel('% Different python-excel (ppm)')
ax2.set_xlabel('SCAS Python (ppm)')

ax3.plot(C_calc['SCAS6_ppm'], C_calc['SCAS6_ppm']-df['SCAS (ppm)'], 'or')
ax3.ticklabel_format(useOffset=False)
ax3.set_ylabel('Python-excel (ppm)')
ax3.set_xlabel('SCAS Python (ppm)')
fig.tight_layout()
```
