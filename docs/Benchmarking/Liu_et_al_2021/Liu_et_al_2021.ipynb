```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import PySulfSat as ss
import Thermobar as pt
pd.options.display.max_columns = None
```

```python
df=ss.import_data('Liu et al., 2021.xlsx', sheet_name='PySulfSat')
```

```python
scss_fixed_Sulf=ss.calculate_Liu2021_SCSS(df=df, T_K=df['T(K)'], P_kbar=df['P(GPa)']*10,
                    Fe_FeNiCu_Sulf=df['XFeS'], H2O_Liq=df['H2O (wt%)'])
scss_fixed_Sulf.head()
```

```python
plt.plot(df['SCSS (ppm)'], scss_fixed_Sulf['SCSS_calc'], 'ok')
plt.plot([500, 5500], [500, 5500], '-r')
plt.xlabel('Spreadsheet SCSS (ppm)')
plt.ylabel('Python SCSS (ppm)')
```
